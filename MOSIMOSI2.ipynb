```python
%matplotlib notebook

import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation



fig, ax = plt.subplots()
ax.set_xlim(0, 50)
ax.set_ylim(0, 20)
line, = ax.plot([], [])



#mempertimbangkan
x = 0
y = 0
g = 9.806
m = 0.15
D = 0.0013

theta = math.radians(35)
sinTheta = math.sin(theta)
cosTheta = math.cos(theta)

Vo = 50
Vx = Vo * cosTheta
Vy = Vo * sinTheta
V = math.sqrt((Vy * Vy) + (Vx * Vx))

ax = -(D/m) * V * Vx
ay = -g-(D/m) * V * Vy

t = 0
deltaT = 0.01

storeX = []
storeY = []


print('============ NUMERIK ============')
while True:
    Vx = Vx + (ax * deltaT)
    Vy = Vy + (ay * deltaT)
    x = x + (Vx * deltaT)
    y = y + (Vy * deltaT)
    t = t + deltaT
    print('y = ',round(y,5))
    print('x = ',round(x,5))
    print('t = ',round(t,2))
    print()
    storeX.append(x)
    storeY.append(y)
    if(y <= 0):
        break


dataX = []
dataY = []
  
def animation_frame(i):
	dataX.append(storeX[i])
	dataY.append(storeY[i])

	line.set_data(dataX, dataY)
	return line, 

animation = FuncAnimation(fig, func=animation_frame, frames=1000, interval=1)
plt.show()


print('============ ANALITIK ============')
x = 0
y = 0

Vx = Vo * cosTheta
Vy = Vo * sinTheta

x = x + (Vx * t) + (0.5 * ax * t * t)
y = y + (Vy * t) + (0.5 * ay * t * t)
print('y = ',round(y,5))
print('x = ',round(x,5))
print('t = ',round(t,2))
```
